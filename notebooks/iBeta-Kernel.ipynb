```python
import matplotlib.pyplot
```

```python
import numpy
```

```python
def ikernel(ws):
    n = len(ws)
    fig, axs = matplotlib.pyplot.subplots(nrows = n, ncols = 2, figsize = (15, 3 * n,), sharey=True)
    
    for _i, w in enumerate(ws):
        i = numpy.linspace(0, w, w)
        i_cov = numpy.var(i)
        i_mu = numpy.mean(i)

        i_z = (i - i_mu) / i_cov

        i_z_norm = i_z * (1 / sum(numpy.abs(i_z)))

        x = list(range(w))
    
        axs[_i][0].bar(x, i_z, label = "beta")
        axs[_i][0].bar(x, i_z_norm, label = "norm")
        
        axs[_i][0].legend()
            
        i_z_shift = i_z - min(i_z)
        i_z_norm_shift = i_z_norm - min(i_z_norm)
    
        axs[_i][1].bar(x, i_z_shift, label = "beta")
        axs[_i][1].bar(x, i_z_norm_shift, label = "norm")
        
        axs[_i][1].legend()
    
```

```python
ikernel([3, 5, 10])
```
